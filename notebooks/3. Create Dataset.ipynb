```python
import os
import pandas as pd
```

```python
BASE_DIR = os.getcwd()
DATA_DIR = os.path.join(BASE_DIR, 'Data')

def create_files(directory):
    files = []
    for x in os.listdir(directory):
        if x.startswith('2'):
            filepath = os.path.join(DATA_DIR, x)
            files.append(filepath)
    files.sort()
    return files

files = create_files(DATA_DIR)
```

```python
def create_df(filepath):
    df = pd.read_csv(filepath)
    df.drop(columns=['Unnamed: 0'], inplace=True)
    df.rename(columns={df.columns[1]: 'appearances'}, inplace=True)
    return df
```

Create dataframes with corresponding names

```python
df_2007 = create_df(files[0])
df_2008 = create_df(files[1])
df_2009 = create_df(files[2])
df_2010 = create_df(files[3])
df_2011 = create_df(files[4])
df_2012 = create_df(files[5])
df_2013 = create_df(files[6])
df_2014 = create_df(files[7])
df_2015 = create_df(files[8])
df_2016 = create_df(files[9])
df_2017 = create_df(files[10])
df_2018 = create_df(files[11])
df_2019 = create_df(files[12])
df_2020 = create_df(files[13])
df_2021 = create_df(files[14])
```

```python
df1 = df_2007.merge(df_2008.merge(df_2009, on='Genre', how='outer'))
df2 = df_2010.merge(df_2011.merge(df_2012, on='Genre', how='outer'))
df3 = df_2013.merge(df_2014.merge(df_2015, on='Genre', how='outer'))
df4 = df_2016.merge(df_2017.merge(df_2018, on='Genre', how='outer'))
df5 = df_2019.merge(df_2020.merge(df_2021, on='Genre', how='outer'))
```

```python
df10 = df1.merge(df2.merge(df3, on='Genre', how='outer'))
df11 = df4.merge(df5, on='Genre', how='outer')
```

```python
df = df10.merge(df11, on='Genre', how='outer')
df = df.fillna('0')
df = df.set_index('Genre')
```

```python
for column in df.columns:
    df[column] = df.loc[:, [column]].astype(int)
```

```python
dataset = df.transpose()
dataset
```

```python
dataset.to_csv(os.path.join(DATA_DIR, 'dataset.csv'))
```
