```python
import os
import pandas as pd
```

```python
BASE_DIR = os.getcwd()
DATA_DIR = os.path.join(BASE_DIR, 'Data')
file_path_1 = os.path.join(DATA_DIR, '2007.csv')
file_path_2 = os.path.join(DATA_DIR, '2008.csv')
file_path_3 = os.path.join(DATA_DIR, '2009.csv')
file_path_4 = os.path.join(DATA_DIR, '2010.csv')
file_path_5 = os.path.join(DATA_DIR, '2011.csv')
file_path_6 = os.path.join(DATA_DIR, '2012.csv')
file_path_7 = os.path.join(DATA_DIR, '2013.csv')
file_path_8 = os.path.join(DATA_DIR, '2014.csv')
file_path_9 = os.path.join(DATA_DIR, '2015.csv')
file_path_10 = os.path.join(DATA_DIR, '2016.csv')
file_path_11 = os.path.join(DATA_DIR, '2017.csv')
file_path_12 = os.path.join(DATA_DIR, '2018.csv')
file_path_13 = os.path.join(DATA_DIR, '2019.csv')
file_path_14 = os.path.join(DATA_DIR, '2020.csv')
file_path_15 = os.path.join(DATA_DIR, '2021.csv')
```

```python
df_2007 = pd.read_csv(file_path_1)
df_2008 = pd.read_csv(file_path_2)
df_2009 = pd.read_csv(file_path_3)
df_2010 = pd.read_csv(file_path_4)
df_2011 = pd.read_csv(file_path_5)
df_2012 = pd.read_csv(file_path_6)
df_2013 = pd.read_csv(file_path_7)
df_2014 = pd.read_csv(file_path_8)
df_2015 = pd.read_csv(file_path_9)
df_2016 = pd.read_csv(file_path_10)
df_2017 = pd.read_csv(file_path_11)
df_2018 = pd.read_csv(file_path_12)
df_2019 = pd.read_csv(file_path_13)
df_2020 = pd.read_csv(file_path_14)
df_2021 = pd.read_csv(file_path_15)
```

```python
df1 = df_2007.merge(df_2008.merge(df_2009, on='Genre', how='outer'))
df2 = df_2010.merge(df_2011.merge(df_2012, on='Genre', how='outer'))
df3 = df_2013.merge(df_2014.merge(df_2015, on='Genre', how='outer'))
df4 = df_2016.merge(df_2017.merge(df_2018, on='Genre', how='outer'))
df5 = df_2019.merge(df_2020.merge(df_2021, on='Genre', how='outer'))
df1 = df1.drop(columns=['Unnamed: 0', 'Unnamed: 0_x', 'Unnamed: 0_y'])
df2 = df2.drop(columns=['Unnamed: 0', 'Unnamed: 0_x', 'Unnamed: 0_y'])
df3 = df3.drop(columns=['Unnamed: 0', 'Unnamed: 0_x', 'Unnamed: 0_y'])
df4 = df4.drop(columns=['Unnamed: 0', 'Unnamed: 0_x', 'Unnamed: 0_y'])
df5 = df5.drop(columns=['Unnamed: 0', 'Unnamed: 0_x', 'Unnamed: 0_y'])
```

```python
df10 = df1.merge(df2.merge(df3, on='Genre', how='outer'))
df11 = df4.merge(df5, on='Genre', how='outer')
```

```python
df = df10.merge(df11, on='Genre', how='outer')
df = df.fillna('0')
df = df.set_index('Genre')
```

```python
for column in df.columns:
    df[column] = df.loc[:, [column]].astype(int)
```

```python
dataset = df.transpose()
dataset
```

```python
dataset.to_csv(os.path.join(DATA_DIR, 'dataset.csv'))
```
