```python
# Setup
BASE_DIR = os.getcwd()
DATA_DIR = os.path.join(BASE_DIR, 'Data')
TEMP_DIR = os.path.join(BASE_DIR, 'Temp')
f1 = os.path.join(TEMP_DIR, '2007.csv')
f2 = os.path.join(TEMP_DIR, '2008.csv')
f3 = os.path.join(TEMP_DIR, '2009.csv')
f4 = os.path.join(TEMP_DIR, '2010.csv')
f5 = os.path.join(TEMP_DIR, '2011.csv')
f6 = os.path.join(TEMP_DIR, '2012.csv')
f7 = os.path.join(TEMP_DIR, '2013.csv')
f8 = os.path.join(TEMP_DIR, '2014.csv')
f9 = os.path.join(TEMP_DIR, '2018.csv')
f10 = os.path.join(TEMP_DIR, '2019.csv')
f11 = os.path.join(TEMP_DIR, '2020.csv')
f12 = os.path.join(TEMP_DIR, '2021.csv')
```

```python
def clean_data(file_path):
    """Clean up scraped data (a single csv file)."""
    # Load csv file
    load = pd.read_csv(file_path)
    # Rename column
    col = load.columns[0]
    year = load.columns[1]
    name = load.rename({col: 'Genre'}, axis=1)
    
    # Create keys and values for a dictionary
    genres = name.Genre.unique().tolist()
    vals = [name.loc[name.Genre == genre].iloc[:, 1].sum() for genre in genres]

    # Create list of dictionaries
    temp = []
    for (genre, val) in zip(genres, vals):
        data = {
            'Genre': genre,
            year: val
        }
        temp.append(data)

    # Create a dataframe from the list of dictionaries
    clean = pd.DataFrame(temp)

    # Eliminate genres that occured less than 10 times
    df = clean.loc[clean[year] > 10]
    csv_path = os.path.join(DATA_DIR, f'{df.columns[1]}.csv')
    df.to_csv(csv_path)
    os.remove(file_path)
    return df
```
